```
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import approximate_taylor_polynomial

# --- 1. Визначення функції та діапазону ---
def f(x):
    return np.cos(4*x) - x + 1

# Діапазон для побудови графіків |x| <= 1
x_range = np.linspace(-1, 1, 400)
y_actual = f(x_range)
x0 = 0 # Точка розкладу

print(f"Функція: f(x) = cos(4x) - x + 1")
print(f"Точка розкладу x0 = {x0}\n")

# --- 2. Побудова полінома Тейлора за допомогою scipy ---
# Використовуємо функцію approximate_taylor_polynomial
degree = 3
taylor_poly_scipy = approximate_taylor_polynomial(f, x0, degree, scale=1.0)
y_taylor_scipy = taylor_poly_scipy(x_range)

print(f"Поліном Тейлора 3-го степеня (scipy):")
print(f"T_3(x) = {taylor_poly_scipy}\n")

# --- 3. Побудова графіка функції та наближення ---
plt.figure(figsize=(10, 6))
plt.plot(x_range, y_actual, label='Оригінальна функція $f(x)$', color='blue', linewidth=2)
plt.plot(x_range, y_taylor_scipy, label='Наближення поліномом Тейлора $T_3(x)$', color='red', linestyle='--', linewidth=2)

# Додавання аналітично знайденого полінома для порівняння (T_3(x) = 2 - x - 8x^2)
y_taylor_manual = 2 - x_range - 8*x_range**2
#plt.plot(x_range, y_taylor_manual, label='T_3(x) (аналітично)', color='green', linestyle=':')


plt.title('Графік функції та її наближення поліномом Тейлора 3-го степеня')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.show()
```
