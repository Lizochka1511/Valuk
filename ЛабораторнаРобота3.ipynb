```
import math

# Задане рівняння (Варіант 1): f(x) = x^3 + 2x^2 - x + 2
def f(x):
    return x**3 + 2*x**2 - x + 2

# Перетворена функція для ітерацій x = phi(x).
# Використовуємо x = phi(x), де phi(x) = x + lambda * f(x)
# Для інтервалу [-3, -2] вибираємо lambda = 0.05
def phi(x):
    # Лямбда підібрана так, щоб забезпечити збіжність на інтервалі
    lambda_val = 0.05
    return x + lambda_val * f(x)

# --- Метод простої ітерації ---
def simple_iteration(x0, iterations_count):
    x_prev = x0
    results = []
    for i in range(iterations_count):
        x_curr = phi(x_prev)
        # Оцінка абсолютної похибки на кожному кроці |x_k - x_{k-1}|
        error = abs(x_curr - x_prev)
        results.append((i + 1, x_curr, error))
        x_prev = x_curr
    return results, x_prev

# --- Виконання завдання ---

# Інтервал відокремлення кореня: [-3, -2]
# Початкове наближення x0 (виберемо середину інтервалу)
x0 = -2.5
num_iterations = 5

print(f"Завдання: Знайти корінь рівняння f(x) = x^3 + 2x^2 - x + 2 = 0")
print(f"Інтервал відокремлення: [-3, -2]")
print(f"Початкове наближення x0 = {x0}, Кількість ітерацій: {num_iterations}\n")

print(f"{'Ітерація':<10} | {'Значення x_k':<15} | {'Похибка |x_k - x_{k-1}|':<25}")
print("-" * 60)

iteration_results, final_root = simple_iteration(x0, num_iterations)

for iter_num, x_val, err in iteration_results:
    print(f"{iter_num:<10} | {x_val:<15.6f} | {err:<25.6e}")

# Оцінка похибки після п'яти ітерацій
final_error = iteration_results[-1][2]

print(f"\n--- Результати ---")
print(f"Наближене значення кореня після {num_iterations} ітерацій: {final_root:.6f}")
print(f"Оцінка похибки після {num_iterations} ітерацій (абсолютна різниця між останніми x): {final_error:.6e}")
```
