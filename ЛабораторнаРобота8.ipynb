```
import numpy as np

# Дані з Таблиці 1
x_points = np.array([2.4, 2.6, 2.8, 3.0, 3.2, 3.4, 3.6, 3.8, 4.0, 4.2, 4.4, 4.6])
y_points = np.array([3.526, 3.782, 3.945, 4.043, 4.104, 4.155, 4.222, 4.331, 4.507, 4.775, 5.159, 5.683])

# Параметри завдання
x_target = 2.4
h = x_points[1] - x_points[0] # Крок h = 0.2
tolerance = 0.001

print(f"Завдання: Обчислити f'(x) та f''(x) при x = {x_target}")
print(f"Крок h = {h}, Точність = {tolerance}\n")

# --- Обчислення таблиці прямих кінцевих різниць (Delta Y) ---

n = len(y_points)
delta_y = np.zeros((n, n))
delta_y[:, 0] = y_points

for j in range(1, n):
    for i in range(n - j):
        delta_y[i, j] = delta_y[i+1, j-1] - delta_y[i, j-1]

# Коефіцієнти, необхідні для формул (перший рядок таблиці різниць)
dy0 = delta_y[0, 1]
d2y0 = delta_y[0, 2]
d3y0 = delta_y[0, 3]
d4y0 = delta_y[0, 4]

# --- Формули для похідних полінома Ньютона для рівновіддалених вузлів ---

# Перша похідна в точці x0 (де q=0)
# f'(x0) = 1/h * (Delta y0 - 1/2 * Delta^2 y0 + 1/3 * Delta^3 y0 - 1/4 * Delta^4 y0 + ...)
f_prime = (1/h) * (dy0 - 0.5 * d2y0 + (1/3) * d3y0 - 0.25 * d4y0)

# Друга похідна в точці x0 (де q=0)
# f''(x0) = 1/h^2 * (Delta^2 y0 - Delta^3 y0 + 11/12 * Delta^4 y0 - ...)
f_double_prime = (1/(h**2)) * (d2y0 - d3y0 + (11/12) * d4y0)

print(f"--- Результати ---")
print(f"f'({x_target}) ≈ {f_prime:.{len(str(tolerance).split('.'))}f}")
print(f"f''({x_target}) ≈ {f_double_prime:.{len(str(tolerance).split('.'))}f}")
```
