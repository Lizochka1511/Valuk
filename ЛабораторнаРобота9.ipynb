```
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import CubicSpline

# Дані з таблиці (Варіант 1)
x_points = np.array()
y_points = np.array()

print(f"Таблиця вихідних даних x: {x_points}")
print(f"Таблиця вихідних даних y(x): {y_points}\n")

# --- 1. Побудова кубічного сплайну ---
# Функція CubicSpline автоматично обчислює всі коефіцієнти та створює об'єкт сплайну
cs = CubicSpline(x_points, y_points, bc_type='natural')

# --- 2. Перевірка роботи сплайну (обчислення значень) ---
# Обчислимо значення сплайну в оригінальних точках (мають збігатися з y_points)
y_check = cs(x_points)
print(f"Перевірка значень сплайну в початкових точках:")
for i in range(len(x_points)):
    print(f"S({x_points[i]}) = {y_check[i]:.4f} (Оригінал: {y_points[i]})")

# Обчислимо кілька проміжних значень для прикладу
x_intermediate = np.array([0.1, 0.3, 0.7, 1.2])
y_intermediate = cs(x_intermediate)
print(f"\nПриклад проміжних значень:")
for i in range(len(x_intermediate)):
    print(f"S({x_intermediate[i]}) = {y_intermediate[i]:.4f}")


# --- 3. Побудова графіка інтерполяційної функції ---

# Діапазон x для гладкого графіка
x_plot = np.linspace(min(x_points), max(x_points), 400)
y_plot = cs(x_plot)

plt.figure(figsize=(10, 6))
plt.plot(x_plot, y_plot, label='Графік кубічного сплайну $S(x)$', color='blue')
plt.scatter(x_points, y_points, color='red', zorder=5, label='Вихідні точки таблиці', marker='o', s=100)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Кубічна сплайн-інтерполяція функції')
plt.legend()
plt.grid(True)
plt.show()
```
