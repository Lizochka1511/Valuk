```
import numpy as np

# Задане рівняння (Варіант 1): f(x) = 9x^4 + 8x^3 + 1.5x^2 + 2x - 10
def f(x):
    return 9*x**4 + 8*x**3 + 1.5*x**2 + 2*x - 10

# --- Метод половинного ділення (Бісекції) ---
def bisection_method(a, b, eps):
    if f(a) * f(b) >= 0:
        return None, 0
    c = a
    iterations = 0
    while (b - a) >= eps:
        c = (a + b) / 2
        if f(c) == 0.0:
            break
        if f(c) * f(a) < 0:
            b = c
        else:
            a = c
        iterations += 1
    return c, iterations

# --- Метод хорд (Хибного положення) ---
def chord_method(a, b, eps):
    if f(a) * f(b) >= 0:
        return None, 0
    x_prev = a
    x_curr = b
    iterations = 0
    while abs(f(x_curr)) >= eps:
        x_next = x_curr - (f(x_curr) * (x_curr - x_prev)) / (f(x_curr) - f(x_prev))
        x_prev = x_curr
        x_curr = x_next
        iterations += 1
    return x_curr, iterations

# --- Виконання завдання ---

tolerance = 0.0001
intervals = [[0.0, 1.0], [-2.0, -1.0]]

print(f"Рівняння: f(x) = 9x^4 + 8x^3 + 1.5x^2 + 2x - 10 = 0")
print(f"Бажана точність (eps): {tolerance}\n")

for i, (a, b) in enumerate(intervals):
    print(f"--- Результати для інтервалу {i+1}: [{a}, {b}] ---")

    # Метод половинного ділення
    root_bisect, iter_bisect = bisection_method(a, b, tolerance)
    if root_bisect is not None:
        print(f"Метод половинного ділення: Корінь = {root_bisect:.5f}, Ітерацій: {iter_bisect}")

    # Метод хорд
    root_chord, iter_chord = chord_method(a, b, tolerance)
    if root_chord is not None:
        print(f"Метод хорд: Корінь = {root_chord:.5f}, Ітерацій: {iter_chord}")

    print("-" * 40)

```
